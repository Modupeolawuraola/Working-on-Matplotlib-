# Matplotlib

Mathplotlib is the most popular library for python , it is gives control of our visiuliazation , its use in place of matlab, for visiuliazation

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0, 5, 11)
y= x**2
```

```python
x
```

```python
y
```

```python
#creating a matplotlib is done by two ways - functional way and obejct oriented method 
#showing the functional method 
plt.plot(x, y)
plt.show()   #showing is as if we are printinmg the cell 
```

```python
#its can also be to show colour of the line 
plt.plot(x, y, 'r-')
```

```python
#adding label to the x and y 
plt.plot(x, y)
plt.xlabel('X-label')
plt.ylabel('Y-label')
plt.title('Title')
```

```python
#putting multiple plot on the same plot 
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(x, y, 'b')
```

```python
#introducing matplotlib object oriented api inedx 
#main idea of using object oriented method is to create a firgure object 
fig = plt.figure()   #creating firgure object 
axes= fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel('X-label')
axes.set_ylabel('Y-label')
axes.set_title('Title')
```

```python
fig= plt.figure()


axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title("Larger Plot")

axes2.plot(y,x)
axes2.set_title("Smaller Plot")
```

# FIRGURE TYPES, AND DPI

```python
Matplotlib allow us to control this aspect 
```

```python
fig= plt.figure(figsize=(3, 2)) #dpi should stay 100 per inches 

ax=fig.add_axes([0, 0,1,1])
ax.plot(x, y)

ax.set_title('Title')
ax.set_ylabel('Y- Label')
ax.set_xlabel('X-Label')
```

```python
#using legend using first method
fig= plt.figure(figsize=(5, 2)) #dpi should stay 100 per inches 

ax=fig.add_axes([0, 0,1,1])
ax.plot(x, x**2)
ax.plot(x, x**3)

ax.set_title('Title')
ax.set_ylabel('Y- Label')
ax.set_xlabel('X-Label')
ax.legend()
```

```python
#using legend using second method
fig= plt.figure(figsize=(5, 2)) #dpi should stay 100 per inches 

ax=fig.add_axes([0, 0,1,1])
ax.plot(x, x**2, label='X-Squared')
ax.plot(x, x**3, label='Y- Cube')

ax.set_title('Title')
ax.set_ylabel('Y- Label')
ax.set_xlabel('X-Label')
ax.legend()
```

```python
#using legend location 1
fig= plt.figure(figsize=(5, 2)) #dpi should stay 100 per inches 

ax=fig.add_axes([0, 0,1,1])
ax.plot(x, x**2, label='X-Squared')
ax.plot(x, x**3, label='Y- Cube')

ax.set_title('Title')
ax.set_ylabel('Y- Label')
ax.set_xlabel('X-Label')
ax.legend(loc=1)   #at the topright coner 
```

```python
#using legend location 2 
fig= plt.figure(figsize=(5, 2)) #dpi should stay 100 per inches 

ax=fig.add_axes([0, 0,1,1])
ax.plot(x, x**2, label='X-Squared')
ax.plot(x, x**3, label='Y- Cube')

ax.set_title('Title')
ax.set_ylabel('Y- Label')
ax.set_xlabel('X-Label')
ax.legend(loc= 10)  #location in the  middle 
```

```python
#wanted to make bigger 
fig= plt.figure(figsize=(8, 2)) #dpi should stay 100 per inches 

ax=fig.add_axes([0, 0,1,1])
ax.plot(x, y)
```

```python
#creating a subplot 
fig, axes = plt.subplots(figsize=(8, 2))

axes.plot(x, y)
```

```python
#creating a subplot, giving the rows, and columns a number 
fig, axes = plt.subplots(nrows=2, ncols= 1, figsize=(8, 2))

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```

# MATPLOTLIB 3

Setting colors, line width, line types in the location in matplotlib

PlotApperances 

```python
fig= plt.figure()

ax= fig.add_axes([0,0,1,1])

ax.plot(x, y, color='#FF8C00', linewidth='10', alpha=0.5)   #RGB HEX COLOUR 

#BASIC COLOURS SUCH AS Purple
#alpha argument can be used with line-with, to chech hoe transparent our width can be 
#lw can be use as short form of linewidth 
```

```python
#creating line style 
fig= plt.figure()

ax= fig.add_axes([0,0,1,1])

ax.plot(x, y, color='#FF8C00', lw= 5, linestyle= '--')   #RGB HEX COLOUR 
```

```python
#creating line style 
fig= plt.figure()

ax= fig.add_axes([0,0,1,1])

ax.plot(x, y, color='#FF8C00', lw= 5, linestyle= '-.')   #RGB HEX COLOUR 
```

```python
#creating line style 
fig= plt.figure()

ax= fig.add_axes([0,0,1,1])

ax.plot(x, y, color='#FF8C00', lw= 5, linestyle= ':')   #RGB HEX COLOUR 
```

```python
#creating line style 
fig= plt.figure()

ax= fig.add_axes([0,0,1,1])

ax.plot(x, y, color='#FF8C00', lw= 1, linestyle='--', marker='o', markersize=20, markerfacecolor= 'blue', 
       markeredgewidth=3, markeredgecolor='yellow')   #RGB HEX COLOUR 

#macros are going to be used when we have just a few datapoint, 
```

```python
#creating line style and different types of markers  
fig= plt.figure()

ax= fig.add_axes([0,0,1,1])

ax.plot(x, y, color='#FF8C00', lw= 1, linestyle= '-.', marker='+')   #RGB HEX COLOUR 
```

```python
fig= plt.figure()

ax= fig.add_axes([0,0,1,1])

ax.plot(x, y, color='#FF8C00', lw= 3, linestyle= '-.', marker='1', markersize= 20)   #RGB HEX COLOUR 
```

```python
#creating line style 
fig= plt.figure()

ax= fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', lw= 1, linestyle='--', marker='o', markersize=20, markerfacecolor= 'green', 
       markeredgewidth=3, markeredgecolor='red') 

ax.set_xlim([0,1])
ax.set_ylim([0,2])

#RGB HEX COLOUR 
```
